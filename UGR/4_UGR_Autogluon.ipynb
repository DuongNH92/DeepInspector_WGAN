```python
import pandas as pd 
import numpy as np 
import sys
from sklearn.preprocessing import MinMaxScaler 
from autogluon.tabular import TabularDataset, TabularPredictor
import warnings
from fastprogress import fastprogress
warnings.filterwarnings("ignore")

#Load a tabular & test dataset:
train_data = TabularDataset('../dataset/UGR/trainnew.csv')
test_data = TabularDataset('../dataset/UGR/test.csv')
print("Balanced dataset UGR'16:")
print(train_data['Label'].value_counts())
print("Test dataset:")
print(test_data['Label'].value_counts())

print("Fit classification models predicting:")
predictor = TabularPredictor(label='Label', eval_metric='accuracy').fit(train_data) #, presets='best_quality') 
```

```python
print("The best model:")
print(predictor.get_model_best())
print("------------------------------------------------")
print(predictor.evaluate(test_data))
```

```python
predictor.fit_summary()
```
